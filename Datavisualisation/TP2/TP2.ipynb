```python
from sklearn.preprocessing import LabelEncoder
Origin_num = LabelEncoder()
```

```python
df_cars['Origin_num'] = Origin_num.fit_transform(df_cars['Origin'])
df_cars.drop(["Origin"],axis='columns',inplace=True)
df_cars
```

```python
plt.figure(figsize=(10,8))
plt.subplot(111)
origins = ['Europe', 'Japan', 'US']
a = plt.scatter(x=df_cars['Horsepower'],y=df_cars['Acceleration'], s=150, c=df_cars['Origin_num'] )
plt.xlabel("Horsepower")
plt.ylabel("Acceleration")
plt.legend(loc='lower left',fancybox=True, shadow=True, ncol=3, fontsize=8, handles=a.legend_elements()[0], labels=origins, title="Origin")
plt.show()
plt.savefig(r'C:\Users\LENOVO\Desktop\cours\S5\Datavisualisation\TP2\horsepower_acceleration_origin-num_scatter')
```
